Gebt die Zeitfunktion x[n] an und skizziert den Funktionsgraphen! 

Hinweis:der Gleichanteil im Zeitbereich ist X [0]/N .

```python
def fourieganz(N, k, n_N, gleichanteil,teiler=1):
    xn = gleichanteil/N + np.cos((2*np.pi)/N * k * n_N)
    #xn = 1/2 + np.cos(2*np.pi/3 *n_N[:4])
    p = 0.05
    pn = np.arange(0, N,p)

    kp = int((k-1)*1/p)+1
    pxn = gleichanteil/N + np.cos((2*np.pi)/N * k * pn)
    plt.subplot(2, 1, 1)
    plt.stem(n_N, xn, 'r', markerfmt='ro')
    plt.plot(pn, pxn, 'b')
    print(gleichanteil/N, "+cos(2pi/", N, "*", k, "*n)")
    from numpy.fft import fft, fftshift
    fxn = gleichanteil/N + np.cos((2*np.pi)/N * k * n_N[0:N:teiler])
    ff= np.abs((fft(fxn,int(N/teiler))))
    plt.subplot(2, 1, 2)
    plt.stem(n_N[:int(N/teiler)], ff, 'b', markerfmt='bo')
```

```python
N = 16
gleichanteil = 8 # x[0]
fs = 8000
k = 5
f_0 = k/N * fs
print("f0 ist ",k,"/",N," *", fs ,"=",f_0,"Hz")
n_N = np.arange(0,N)

fourieganz(N, k, n_N, gleichanteil)
```
