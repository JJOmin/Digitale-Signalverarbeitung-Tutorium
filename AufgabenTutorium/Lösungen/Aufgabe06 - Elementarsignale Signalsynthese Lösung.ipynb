## [Sprungfunktion](https://de.wikipedia.org/wiki/Heaviside-Funktion)

1. Plotte einen einfachen Sprung für Wertebereich von -5 bis 5 , also wenn x[n] = σ[n]
    - Hinweis: np.heavside & np.arange werden für die berechnung benötigt

```python
import numpy as np
import matplotlib.pyplot as plt
def sprung(n):
    return np.heaviside(n,1)

n = np.arange(-5,5+1) #Wertebereiches des Signals
signal = sprung(n)

#Plotten des Signals
plt.title("Darstellung im Zeitbereich des Sprungsignals  x[n] = σ[n]")
plt.stem(n,signal)
plt.grid(True)
```

2. Plotte 2 Sprünge im Wertebereich -5 bis 5,
- Einer der um 2 nach rechts verschoben ist
- Einer der um 3 nach links verschoben ist

```python
import numpy as np
import matplotlib.pyplot as plt
def sprung(n):
    return np.heaviside(n,1)

n = np.arange(-5,5+1) #Wertebereiches des Signals

x2 = sprung(n-2) #Verschiebung des Signals um 2 Einheiten nach rechts
x3 = sprung(n+3) #Verschiebung des Signals um 2 Einheiten nach rechts


#Plotten des Signals
plt.figure(figsize=(10, 12))
plt.subplot(4, 2, 1)
plt.title("x2[n] = σ[n-2]")
plt.stem(n,x2)
plt.grid(True)

plt.subplot(4, 2, 2)
plt.title("x3[n] = σ[n+3]")
plt.stem(n,x3)
plt.grid(True)
```

## [Dirac-Funktion/Impuls](https://de.wikipedia.org/wiki/Delta-Distribution)

- Elementarsignale - Plotte einen Standardimpuls für Wertebereich von -5 bis 5  x4[n] = δ[n]

```python
def dirac(n):
    return np.array([1 if 0 == i else 0 for i in n]) #Impulsfunktion

n = np.arange(-5,5+1) #Wertebereiches des Signals
x4 = dirac(n) #Verschiebung des Signals um 1 Einheit nach rechts

#Plotten des Signals
plt.title("Darstellung Zeitbereich des Impuls für x4[n] = δ[n]")
plt.stem(n,x4)
plt.grid(True)
```

- Plotte <b>x5[n] = 5δ[n + 1] + 5</b> für Wertebereich von -5 bis 5  

```python
def dirac(n):
    return np.array([1 if 0 == i else 0 for i in n]) #Impulsfunktion

n = np.arange(-5,5+1) #Wertebereiches des Signals

x5 = 5*dirac(n+1)+5 #Verschiebung des Signals um 1 Einheit nach rechts

#Plotten des Signals
plt.title("Darstellung im Zeitbereich des Impuls x5[n] = 5δ[n+1]")
plt.stem(n,x5)
plt.grid(True)
```

## Signalsynthese von Sprung & Sprung

- Plotte <b>x6 = 3σ[n-3] + 2σ[n+3]</b> für Wertebereich von -5 bis 5  

```python
def sprung(n):
    return np.heaviside(n,1)

n = np.arange(-5,5+1) #Wertebereiches des Signals

x6 = 3* sprung(n-3) + 2* sprung(n+3)  #Verschiebung des Signals um 1 Einheit nach rechts

#Plotten des Signals
plt.title("Signalsynthese von Sprung & Sprung x6 = 3δ[n-3] + 2δ[n+3]")
plt.stem(n,x6)
plt.grid(True)
```

## Signalsynthese von Dirac & Dirac

- Plotte x7 = -3*δ[n + 2] + 2*σ[n-3] für -5 bis 5

```python
def dirac(n):
    return np.array([1 if 0 == i else 0 for i in n]) #Impulsfunktion

n = np.arange(-5,5+1) #Wertebereiches des Signals

x7 = -3* dirac(n+2) + 2* dirac(n-3)  #Verschiebung des Signals um 1 Einheit nach rechts

#Plotten des Signals
plt.title("Dirac & Dirac x7 = -3*δ[n + 2] + 2*σ[n-3]")
plt.stem(n,x7)
plt.grid(True)
```

## Signalsynthese von Sprung & Dirac

- Plotte x8 = δ[n - 8] + σ[n-3] für -10 bis 10

```python
def dirac(n):
    return np.array([1 if 0 == i else 0 for i in n]) #Impulsfunktion

def sprung(n):
    return np.heaviside(n,1)

n = np.arange(-10,10+1) #Wertebereiches des Signals

x8 = dirac(n-8) + sprung(n-3)  #Verschiebung des Signals um 1 Einheit nach rechts

#Plotten des Signals
plt.title("Signalsynthese von Impuls und Dirac x8 = δ[n - 8] + σ[n-3]")
plt.stem(n,x8)
plt.grid(True)
```
