# Sprungfunktion

```python
import numpy as np
import matplotlib.pyplot as plt

#Standardwerte Sygnalsysnthes
n = np.arange(-10,16+1) #Wertebereiches des Signals
print(n)
array = np.heaviside(n+5,1)*2
print(array.size)
plt.stem(n,array) #Plot des arrays, über alle werte in n


```

```python
n = np.arange(0,15) #Was passiert hier?
print(np.heaviside(n,1)) #Was wird hier returnt?
```

```python
#Rechenoperationenmit NuzmPy Arrays

print(array @ array) # Skalarprodukt
print(array * array) # Multiplikation

print(array.size) #Anz. elemente
print(array.ndim) #Dimension des Arrays
print(array.sum()) #Summe aller Elemente im Array
```
